# Exchange Rates Data Visualization

The purpose of this analysis project is to explore the ways in which the matplotlib fivethirtyeight style can be deployed to create informative, parsimonious data visualizations. The goal is to minimize clutter and unnecessary aspects of plots, while highlighting the relevant trends and messages. 

## Data

The data for this project reflects the Euro exchanges rates across the European Union and other world banks, and is freely available on [Kaggle](https://www.kaggle.com/datasets/lsind18/euro-exchange-daily-rates-19992020). The data is sourced from the European Central Bank and is regularly updated. For this project, the data was taken from January 2021. 

```python
#importing pandas library
import pandas as pd
#read in the data to a dataframe
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

### Data inspection and cleaning

```python
exchange_rates.info()
```

The data set contains 5699 (daily) entries reflecting exchanges rates from 40 different currencies across the globe. Many of the columns contain null values, and most columns have data stored as `object` datatypes rather than integers or floats. 

The first pass of this data will focus on the Euro-to-US dolar exchange rate, so I will isolate that data first. 

```python
exchange_rates.rename(columns = {'[US dollar ]' : 'US_dollar',
                                 '[Swiss franc ]' : 'Franc',
                                 'Period\\Unit:' : 'Time'},
                      inplace = True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace = True)
exchange_rates.reset_index(drop = True, inplace = True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar','Franc']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
#Remove all US_dollar values with a '-' character and convert data type
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar = euro_to_dollar[euro_to_dollar['Franc'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar['Franc'] = euro_to_dollar['Franc'].astype(float)
euro_to_dollar.info()
```

## Data visualization

### Euro to US Dollar

One valuable aspect of this data set is that it is collected daily over many years, this allows us to view what is called a time series - or a series of data points sequenced over time (T). Time series data can be noisy if the $\Delta$T is relatively small. In the plot below, note how the line is not smooth, but rather ragged, indicating the small daily fluctuations in rate in addition to the broader changes in the exchange rate over time. 

```python
import matplotlib.pyplot as plt
%matplotlib inline 

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

If we are not particularly interested in the daily fluctuation of rate, we could instead use a rolling average over a time scale that makes sense for our visualization. 

Below, I've plotted examples of the data plotted as rolling means of various sizeed windows. Observe how as the window increases, the data smooths. 

```python
plt.figure(figsize= (9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original values', weight = 'bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6], [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], 
            euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling window: ' + str(rolling_mean), weight = 'bold')
    
plt.tight_layout()
plt.show()
```

As the data smooths in larger rolling windows, the low-level noise is erased, but if the window is very large, some of the other features of the data are erased as well. It is important to select a rolling window that preserves the trends of interest to your analysis, while smoothing some of the less relevant noise. For this data, I've chosen a rolling window of 30, reflecting the monthly trends in data. 

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar['rolling_mean_f'] = euro_to_dollar['Franc'].rolling(30).mean()
euro_to_dollar
```

### Euro to US Dollar during three different American presidencies

This section gives a visualization of how exchange rates between the Euro and the US dollar fared during the Bush (2001-2008), Obama (2009-2016), and Trump (2016-2020) presidencies. 

```python
b_o_t = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2001) & 
                             (euro_to_dollar['Time'].dt.year < 2021)]
bush = b_o_t.copy()[b_o_t['Time'].dt.year < 2009]
obama = b_o_t.copy()[(b_o_t['Time'].dt.year >=2009) & 
                     (b_o_t['Time'].dt.year < 2017)]
trump = b_o_t.copy()[(b_o_t['Time'].dt.year >= 2017) &
                     (b_o_t['Time'].dt.year < 2021)]

averageExchange = b_o_t['US_dollar'].mean()
averageExchange
```

### Euro to US Dollar during the 2007-2008 financial crisis

This section gives a visualization of how the US dollar fared against the Euro during the financial crises of 2007-2008 in the US. 

```python
#Isolating the time around the crisis, and the crisis years specifically
fcrisis = euro_to_dollar.copy(
            )[(euro_to_dollar['Time'].dt.year >= 2006
            ) & (euro_to_dollar['Time'].dt.year <= 2009)]
fcrisis78 = euro_to_dollar.copy(
            )[(euro_to_dollar['Time'].dt.year >= 2007
            ) & (euro_to_dollar['Time'].dt.year <=2008)]
peak = fcrisis['US_dollar'].max()
peak
```

```python
# Forming the plot
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(fcrisis['Time'], 
       fcrisis['rolling_mean'],
       linewidth = 1, color ='#a6d785')

# Highlighting the 2007-2008 period
ax.plot(fcrisis78['Time'], 
       fcrisis78['rolling_mean'],
       linewidth = 3, color = '#6d904f')

# Highlighting the peak of the crisis
ax.axvspan(xmin = 733112.0, xmax = 733302.0, ymin = 0.09, 
           alpha = 0.3, color = 'grey')

# Adding tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 732272.0
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, 1.13, year, alpha = 0.5, fontsize = 11)
    x += 365
    
y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5']: 
    ax.text(732172.0, y, rate, alpha = 0.5, fontsize = 11)
    y += 0.1
    
# Adding a title and subtitle
ax.text(732172.0, 1.67, 'Euro-USD rate peaked at 1.60 during the 2007-2008 financial crisis in the US', 
        weight = 'bold')
ax.text(732172.0, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
       size = 12)

# Adding grid transparency
ax.grid(alpha = 0.5)
plt.show()
```

### Euro to US Dollar, Swiss Franc, and Singaporean Dollar during the coronavirus pandmic

This final section depcits how the outbreak of the SARS-CoV-2 pandemic affected exchange rates in for the US Dollar, the Swiss Franc, and the Singaporean dollar. 

```python
#Isolating the time leading up to the pandemic, and the pandeimc specifically
pandy = euro_to_dollar.copy(
            )[euro_to_dollar['Time'] >= '2019-09-01']
pandy_US = euro_to_dollar.copy(
            )[euro_to_dollar['Time'] >= '2020-03-15']
```

```python
# Forming the plot
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(pandy['Time'], 
       pandy['rolling_mean'],
       linewidth = 1, color ='#8B8386')
ax.plot(pandy_US['Time'], 
       pandy_US['rolling_mean'],
       linewidth = 2.5, color ='#8b8b8b')

ax.plot(pandy['Time'], 
       pandy['rolling_mean_f'],
       linewidth = 1, color ='#ff83fa')
ax.plot(pandy_US['Time'], 
       pandy_US['rolling_mean_f'],
       linewidth = 2.5, color ='#cd00cd')

# Identifying Declaration of pandemic
fnt = 8
txtc = '#27408B'
ax.vlines(x = 737432, ymin = 1.05, ymax = 1.18, color =txtc,
          linewidth = 1, linestyle = 'dashed')
ax.text(737432, 1.21, 'WHO declares', color = txtc,
        fontsize = fnt, horizontalalignment = 'center')
ax.text(737432, 1.195, 'COVID-19 pandemic', color = txtc,
       fontsize = fnt, horizontalalignment = 'center')
ax.text(737432, 1.03, 'Jan 10', color = txtc,
       fontsize = fnt+1, horizontalalignment = 'center')

# Identifying when shutdowns begin across US and Europe
ax.vlines(x = 737500, ymin = 1.04, ymax = 1.14, color =txtc,
          linewidth = 1, linestyle = 'dashed')
ax.text(737500, 1.165, 'Shutdowns begin', color = txtc,
        fontsize = fnt, horizontalalignment = 'center')
ax.text(737500, 1.15, 'in US and Europe', color = txtc, 
       fontsize = fnt, horizontalalignment = 'center')
ax.text(737500, 1.02, 'Mar 15', color = txtc,
       fontsize = fnt+1, horizontalalignment = 'center')
              
              
# Setting tick labels
ax.set_xticklabels(['', "Nov 2019", '' ,"Mar 2020",
                     '', "Jul 2020" ,'', "Nov 2020", ''],
                   alpha = 0.5, fontsize = 12)

ax.set_yticks([1.0, 1.1, 1.2, 1.3])
ax.set_ylim(0.98, 1.25)
ax.set_yticklabels(['1.0','1.1', '1.2','1.3'], alpha = 0.3)
ax.grid(alpha = 0.3)

# Adding titles/labels
# Adding labels
ax.text(737700, 1.15, 'USD', color = '#8b8b8b',
       fontsize = 14, weight = 'bold')
ax.text(737700, 1.05, 'Franc', color = '#cd00cd',
       fontsize = 14, weight = 'bold')
ax.text(737250.0, 1.27, 'COVID-19 dropped EURO-USD, EURO-Franc exchange rates', 
         fontsize = 16, weight = 'bold')
ax.text(737250.0, 1.24, '''USD rates climbed over Summer 2020 while the Swiss franc stagnated''',
         fontsize = 14)
plt.show()
```

The above graphic depicts the way declaration of a worldwide pandemic and subsequent shut downs impacted EURO exchange rates with the US and with Switzerland. Following the WHO's declaration, the Swiss Franc began a steady decline that did not begin to recover until well into the Summer of 2020, and even then, recovery was slow. Conversely, the US lagged behind Europe in responding to the crisis as well as in the affects on the exchange rate, but also recovered more speedily and robustly over the ensuing months.
